```python
def plot_decision_boundary(clf, X, y, ax):
    ax.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.jet, edgecolors='k')

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300),
                         np.linspace(y_min, y_max, 300))

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot decision regions    ax.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.jet)
```

```python
lin_svc = LinearSVC(loss='hinge', multi_class='crammer_singer')
c_vals = [0.01, 0.1, 1.0, 10.0]
X = df.iloc[:, :-1].values
#print(X)
y = df.iloc[:, -1].values
#print(y)

fig, axes = plt.subplots(2, 2, figsize=(10, 10))
axes = axes.ravel()

for i, c_val in enumerate(c_vals):
    lin_svc = LinearSVC(loss='hinge', multi_class='crammer_singer', C=c_val)
    lin_svc.fit(X,y)
    plot_decision_boundary(lin_svc, X, y, axes[i])
    axes[i].set_title(f"C = {c_val}")

plt.tight_layout()
plt.show()
```
